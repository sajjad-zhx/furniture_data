```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_furniture_data.csv')
df.head()
```

```python
# Summary of numeric columns
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot price distribution
sns.histplot(df['price'], kde=True)
plt.title("Distribution of Product Prices")
plt.xlabel("Price ($)")
plt.ylabel("Number of Products")
plt.show()
```

```python
sns.histplot(df['sold'], kde=True, color='orange')
plt.title("Distribution of Items Sold")
plt.xlabel("Number of Items Sold")
plt.ylabel("Number of Products")
plt.show()
```

```python
sns.scatterplot(x='price', y='sold', data=df)
plt.title("Price vs Items Sold")
plt.xlabel("Price ($)")
plt.ylabel("Items Sold")
plt.show()
```

```python
sns.countplot(x='tagText', data=df)
plt.title("Shipping Type Count")
plt.xlabel("Shipping Type")
plt.ylabel("Number of Products")
plt.show()
```

```python
sns.boxplot(x='tagText', y='price', data=df)
plt.title("Price Distribution by Shipping Type")
plt.xlabel("Shipping Type")
plt.ylabel("Price ($)")
plt.show()
```

```python
df.groupby('tagText')['sold'].mean()
```

```python
# Only use numeric columns
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Histogram of product prices
sns.histplot(df['price'], kde=True)
plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# Final price distribution
sns.histplot(df['final_price'], kde=True, color='green')
plt.title("Distribution of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Count")
plt.show()
```

```python
# See if some prices are extremely high
sns.boxplot(x=df['price'])
plt.title("Outliers in Product Price")
plt.show()
```

```python
sns.pairplot(df[['price', 'shipping_price', 'sold']])
plt.suptitle("Relationships Between Numeric Features", y=1.02)
plt.show()
```
